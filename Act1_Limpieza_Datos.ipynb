**Parte 1:** Las bases de datos (BD) son bancos de información, colecciones o listas que se encargan de almacenar diferentes tipos de datos y conectarlos entre sí bajo una unidad de organización lógica. Está información se puede recopilar desde diferentes orígenes, puede ser compartida y consultada por diferentes aplicaciones o softwares.
Para la ciencia de datos la información almacenada en las bases de datos sirve para realizar análisis de información que posibilita la toma de decisiones. Para lograr el análisis de información basado en datos, se debe seguir un proceso iterativo de ciencia de datos:

1.	Definición del problema
2.	Recolección de datos y preparación
3.	Desarrollo e implementación de un modelo
4.	Mejora del rendimiento

Las BD deben pasar por un proceso de limpieza y pre-procesamiento exhaustivo en el que intervienen los científicos de datos para conseguir un conjunto de datos de calidad, con información relevante, validada y que modele de forma correcta un comportamiento real, permitiendo la obtención de los mejores resultados al utilizar estos datos en un modelo.
El preprocesamiento de forma resumida consiste en transformar datos incomprensibles en comprensibles, en darle sentido a datos con discrepancias o errores y en llevar a cabo la limpieza, transformación y reducción de datos.
Por otro lado, los almacenes de datos o Data Warehouse para ciencia de datos se caracterizan por ser un depósito centralizado de la información a analizar que en general son accedidos mediante softwares de inteligencia empresarial, otras fuentes de BD y otras aplicaciones.
